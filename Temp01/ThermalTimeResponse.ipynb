Figure: Tubular ceramic encased Pt100 elements. The aim is to estimate the time response

of fused silica ceramic material is about ùúå = 2.2

We also assume that we can neglect the platinum filament inside the sensor. In

coming out from the sensor. <br>
Figure: Wire-wound PRT. The platinum filament is coiled around the ceramic insulator and 
covered with ceramic shield (see element types in resistance thermometers). For thermal 
analysis we simplify and assume that the sensor element is made only of ceramic material.<br>
Estimate
- the total volume ùëâ
- the total mass ùëö, and
- the outer surface area of the ceramic element.<br>
Let‚Äôs assume that the heat transfer coefficient for an object in hot water is about 
‚Ñé = 1000 ùëä
ùëö2ùêæ. <br>
Based on all these calculations 
- Estimate the thermal rate for the sensor.
- Draw the time response curve for this sensor.
- Based on the curve what is the sensor‚Äôs temperature at ùë° = 10.0 ùë†.

```python
To = 22
Tenv = 100
density = 2.2 * 1000 #g/cm3 into kg/m3
Cp = 740
h = 1000 # heat transfer co-efficient
#The ceramic tube's dimensions, from the datasheet
diameter = 5 / 1000 # in meters
radius = diameter / 2
length = 35 / 1000 
area = 2 * pi * radius * (radius + length)
volume = pi * (radius)**2 * length
m = density * volume #mass
r = (h*A)/(Cp*m)
t = arange(0, 30, 0.5)
T = Tenv + (To -Tenv)*exp(-r*t)
T10 = Tenv + (To -Tenv)*exp(-r*10)
figure(figsize=(10, 6))
plot(t, T)
title('Time response curve for pt100')
xlabel('Time (s)')
ylabel('Temperature (C)')
grid()
print('\nThe volume is', round(volume, 8), 'cubic meters, mass is', round(m, 5), 'kg, and the surface area is', round(area, 5), 'square meters\n')
print('The thermal rate for the sensor is', round(r, 3),'\n')
print('The sensors temperature at t = 10s is over 99 degree centigrade (but it is',round(T10, 2), 'as per the calculation)')
```
