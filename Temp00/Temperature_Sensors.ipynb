2. The second-order polynomial approximation for the resistance of Pt100 (having accuracy better than 0.01 [°C]) can be given by equation: 𝑅(𝑇) = 𝑅0(1 + 39.08 × 10−4 ⋅ 𝑇 + 5.8 × 10−7 ⋅ 𝑇^2), where 𝑅0 is the nominal resistance.
  
a. Make tables and sketch graphs of the linear and second order models for Pt100 within temperature range from −40 [°C] to +120 [°C].   
b. How much do the first order model differ from the second order approximation in the given temperature range?    
c. What is the maximum difference in resistance and on which temperature does it occur?

```python
t= np.arange(-40, 140, 20);
r = 100+0.385*t + 0.000058* (t**2)
print('T =', t, ', R =', r)
plt.plot(t,r)
plt.ylabel("Resistance(Ω)")
plt.xlabel("Temperature[°C] ")
show()
```

```python
r1 = 100+0.385*t
r2 = 100+0.385*t + 0.000058* (t**2)
y= r2-r1
print('Differenc between two first order and second order =', y)
max = np.max(y)
print('Maximum difference =', "{:.5f}".format(max))
plt.figure(figsize=(10,10))
plt.plot(t,r1)
plt.plot(t,r2)
plt.ylabel("Resistance(Ω)")
plt.xlabel("Temperature[°C] ")
show()
```

Maximum resistance difference occured at 120 ohm.
